#### Slicing and Reshaping Arrays

* Slicing helps extract parts of arrays.
* Reshaping changes the shape without changing data

```python
import numpy as np

arr = np.arange(12)
print("Original:", arr)

#Slicing
print("First 5:", arr[:5])
print("Last 3:", arr[-3:])
print("from 3rd to seventh:",arr[3:12:3])
```

```python
#Reshape to 3x4
reshaped=arr.reshape(3,4)
print("Reshaped 3x4:\n",reshaped)
print("\nLast column:\n", reshaped[:,-2])
print("\nLast row:\n",reshaped[-1:])
reshaped[0][-1]=55
```

##### Boolean Indexing & Filtering

- Use boolean masks to filter arrays based on conditions.

- In NumPy, you can pass a boolean array inside square brackets to filter the original array. This is called boolean indexing.

```python
scores = np.array([55, 72, 89, 45, 99, 66, 78])
high_scores = scores[scores > 40]
print("Scores > 70:", high_scores)
#find even scores
even_score=scores[scores%2==0]
print("even_score:", even_score)
odd_score=scores[scores%2==1]
print("odd_score:", odd_score)
```

## 3. Working with Missing Data in NumPy

Unlike **Pandas**, standard **NumPy integer arrays** do not support missing values (`NaN`).  
To work with missing data, the array usually needs to be of type **float** or **object**.

### ðŸ”¹ Key Concepts

- **`np.nan`**  
  Represents a missing or undefined value (*Not a Number*).

- **`np.isnan()`**  
  Checks which elements are missing.

- **`np.nansum()`**, **`np.nanmean()`**  
  Perform arithmetic operations while safely ignoring `NaN` values.

```python
# Create an array with missing data (must be float)
data = np.array([1.0, 2.0, np.nan, 4.0, 5.0, np.nan])

print(f"Array with NaNs: {data}")
# Detect missing values
bool_miss = np.isnan(data)
print(f"Is NaN?: {bool_miss}")

# Count missing values
print(f"Count of NaNs: {np.sum(bool_miss)}")
```

```python
# 3. Standard aggregation fails with NaN
print(f"Standard Sum (fails): {np.sum(data)}")


# 4. Safe aggregation (ignoring NaNs)
print(f"Safe Sum (nansum): {np.nansum(data)}")
print(f"Safe Mean (nanmean): {np.nanmean(data)}")
```

```python
# 5. Removing NaNs
clean_data = data[~np.isnan(data)]
print(f"Data with NaNs removed: {clean_data}")

# 6. Filling NaNs with a value (e.g., 0)
filled_data = np.copy(data)
filled_data[np.isnan(filled_data)] = 0
print(f"Data with NaNs filled with 0: {filled_data}")
```
